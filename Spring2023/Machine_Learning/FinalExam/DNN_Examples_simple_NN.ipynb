## Exercise

Make and visualize a NN with the following structure:

An input layer with 3 neurons
One hidden layer with 9 neurons
An output layer with 4 neurons

```python
model = Sequential()# Build the model
model.add(Input(3))# Input Layer
model.add(Dense(9))# Hidden Layers
model.add(Dense(4))# Output layer
visualizer(model, file_name="model", file_format='png')
Image("model.png")# Visualizing the network
```


Sequential(...) can also take a list of layers that you want to add.

```python
model = Sequential([Input(3), Dense(9), Dense(4)])

# Visualizing the network
visualizer(model, file_name="model", file_format='png')
Image("model.png")
```

## Exercise

Create and visualize a new model with 1 hidden layer.

The input layer should have 2 neurons

The hidden layer should have 3 neurons with activation function: "relu"

The output layer should have 4 neurons with activation function: "relu"

```python
model = Sequential()# Build the model.
model.add(Input(2))# Input Layer
model.add(Dense(3, activation="linear"))# Hidden Layer
model.add(Dense(4, activation="relu"))# Hidden Layer
model.add(Dense(3, activation="sigmoid"))# Output Layer
visualizer(model, file_name="model", file_format='png')
Image("model.png")# Visualizing the network
```

Create and visualize a new model with three hidden layers.

The input layer should have 15 neurons

The first hidden layer should have 8 neurons with activation function : "sigmoid"

The second hidden layer should have 6 neurons with activation function : "linear"

The third hidden layer should have 4 neurons with activation function : "relu"

The output layer should have 4 neurons

```python
# Build the model.
model = Sequential()

# Input Layer
model.add(Input(15))

# Hidden Layer
model.add(Dense(8, activation="sigmoid"))
model.add(Dense(6, activation="linear"))
model.add(Dense(4, activation="relu"))

# Output Layer
model.add(Dense(4))

# Visualizing the network
visualizer(model, file_name="model", file_format='png')
Image("model.png")
```
