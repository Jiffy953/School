```python
cars_numeric = cars.select_dtypes(include=['float', 'int'])
cars_numeric.info()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
cars_numeric.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
y = cars_numeric["price_usd"]
cars_numeric.drop(columns="price_usd", inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_cars_numeric_array = scaler.fit_transform(cars_numeric)
scaled_cars_numeric = pd.DataFrame(scaled_cars_numeric_array, index=cars_numeric.index, columns=cars_numeric.columns)
scaled_cars_numeric
```

```python
cars.groupby("manufacturer_name")["price_usd"].mean().sort_values(ascending=False)
```

```python
cars.groupby("body_type")["price_usd"].mean().sort_values(ascending=False)
```

```python
cars["price_range"] = pd.cut(cars["price_usd"], [0,10000,25000,50000], labels=['Cheap', 'Medium', 'Expensive'])
cars["price_range"] 
```

```python
cars_dummies = pd.get_dummies(cars["body_type"], prefix="body_type", prefix_sep='_')
cars_dummies
```

```python
tempdf = pd.get_dummies(cars["engine_fuel"], prefix="engine_fuel", prefix_sep='_')
cars_dummies = pd.concat([cars_dummies, tempdf])
X = pd.concat([cars_dummies, scaled_cars_numeric])
X
```

```python
temp = pd.DataFrame(pd.concat([X, y]))
temp.corr()
```

```python
X = X[["odometer_value", "number_of_photos", "duration_listed", "up_counter", "engine_capacity"]]
X=X.dropna()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

```

```python
from sklearn import datasets, linear_model

regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = regr.predict(X_test)
mean_squared_error(y_test, y_pred, squared=False)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=4, include_bias=False)
X_poly = poly_features.fit_transform(X)
X_poly.shape
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.2)
```

```python
polyregr = linear_model.LinearRegression()
polyregr.fit(X_train_poly, y_train)
y_poly_pred = polyregr.predict(X_test_poly)
mean_squared_error(y_test, y_poly_pred, squared=False)
```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
from sklearn.pipeline import make_pipeline
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier


bc = datasets.load_breast_cancer()

rfc = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=0)
train_sizes, train_scores, test_scores = learning_curve(rfc, X, y, cv=10, train_sizes=np.linspace(0.1, 1.0, 10),n_jobs=1)


train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training Accuracy')
plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean, color='green', marker='+', markersize=5, linestyle='--', label='Validation Accuracy')
plt.fill_between(train_sizes, test_mean + test_std, test_mean - test_std, alpha=0.15, color='green')
plt.title('Learning Curve')
plt.xlabel('Training Data Size')
plt.ylabel('Model accuracy')
plt.grid()
plt.legend(loc='lower right')
plt.show()


```
