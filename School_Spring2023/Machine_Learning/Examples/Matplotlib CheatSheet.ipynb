```python
fig, ax = plt.subplots()
x = np.random.randint(1,10, size=10)
y = 2*x
plt.plot(x,y) # same as ax.plot(x,y)
plt.show()
#See for more about plt.plot parameters and examples: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
```

```python
# Multiple lines
fig, ax = plt.subplots()
x = np.random.randint(1,10, size=10)
y = 2*x
plt.plot(x,y)
plt.plot(x,x+3)
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.linspace(0,10,1000)
y = 2*x
# set of default color and style
plt.plot(x, np.cos(x))
# RGB tuple, values 0 to 1, solid style
plt.plot(x, np.sin(x), color=(1.0,0.2,0.3),linestyle='-')
# specify color by name, dashed style
plt.plot(x, y, color='blue', linestyle='--')
# short color code (rgbcmyk), dotted style
plt.plot(x, x+3, color='g',linestyle=':')
# Grayscale between 0 and 1, dashdot style
plt.plot(x, np.log(x+0.001), color='0.75',linestyle='-.')
# Hex code (RRGGBB from 00 to FF), dashed style
plt.plot(x, x, color='#FFDD44',linestyle='--')
plt.plot(x, np.sqrt(x), color='orange',linestyle='--')
plt.show()
```

```python
plt.xlim(0, 11) #Get or set the x limits of the current axes.
plt.ylim(-1.5, 10) #Get or set the x limits of the current axes.
print(plt.xlim())
print(plt.ylim())
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, np.sin(x), label='y=sin(x)')
plt.plot(x,np.cos(x), label='y=cos(x)')
plt.title('Sine and Cosine Functions ')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend() # Describing the element of the graph
plt.show()
```

## Visualization on World Happiness Index

```python
import pandas as pd
df = pd.read_csv('happiness_rank.csv')
df.head(3)
```

__Scatter plot__
A scatter plot consists of multiple data points plotted across two axes. Each variable depicted in a scatter plot would have multiple observations. This can be a very useful chart type whenever we would like to see if there is any relationship between two sets of data.

```python
fig, ax = plt.subplots(figsize = (12,6))
#**fig_kw All additional keyword arguments are passed to the pyplot.figure call.
x = df['GDP']
y = df['Score']
plt.scatter(x,y, label='y=sin(x)')
#plt.plot(x,y)
#plt.bar(x,y)
plt.title('GDP vs Happiness Score')
plt.xlabel('GDP')
plt.ylabel('Score')
plt.legend()
plt.show()
```

__Bar Chart__
Bar charts are among the most frequently used chart types. As the name suggests, a bar
chart is composed of a series of bars illustrating a variable’s development.
There are four types of bar charts: regular bar chart, verticle bar chart, group bar chart,
and stacked bar chart.

```python
# Prepare Data
countries = ['United States','Japan', 'Germany','Brazil', 'India']
y_pos = np.arange(len(countries))
df_19 = df.loc[df['Year'] == 2019]
data = df_19[(df_19['Country'].isin(countries))].sort_values(['Country'])
data.sort_values('GDP', inplace=True)
data.reset_index(drop=True)

# Bar Chart - remvoe comment to see the effect
plt.bar(y_pos, data['GDP'], align='center', alpha=0.5)
#plt.xticks(y_pos, data['Country'])
#plt.ylabel('GDP')
#plt.title('Bar Chart')
plt.show()
```

```python
# Horizontal bar chart
df_19[(df_19['Country'].isin(countries))].sort_values(['Country'])
data.sort_values('Score', inplace=True)
data.reset_index(drop=True)
plt.barh(y_pos, data['Score'], align='center', alpha=0.5)
plt.yticks(y_pos, data['Country'])
plt.xlabel('Score')
plt.title('Horizontal Bar Chart')
plt.show()
```

```python
# Group bar chart
# matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)
# See for more: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar

index = np.arange(5)
width = 0.3
fig, ax = plt.subplots(figsize=(9, 6))
rects1 = ax.bar(index - width / 2, data['GDP'], width, color='#1f77b4', alpha=0.5)
rects2 = ax.bar(index + width / 2, data['Health'], width, color='#1f77b4')
plt.xticks(index, data['Country'])
plt.legend((rects1[0], rects2[0]), ('GDP', 'Health'))
plt.show()
```

__Histogram__
A histogram is a vertical bar chart that depicts the distribution of a set of data.
Histograms are used to show distributions of variables while bar charts are used to
compare variables. Histograms plot quantitative data with ranges of the data grouped
into bins or intervals while bar charts plot categorical data.

Bar graphs have space between the columns, while histograms do not.
Histograms are great when we would like to show the distribution of the data we
are working with. This allows us to group continuous data into bins and hence, provide a
useful representation of where observations are concentrated.

```python
df_19['Corruption'] #ranges between 0 and 1
```

```python
# Histogram
# See for more: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
# matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs)
fig = plt.figure(figsize=(8,6))
plt.hist(df_19['Corruption'], bins=25, density=True)
plt.grid(alpha=0.5)
plt.show()
```

__Time Series Plot__
Time-series graphs can be used to visualize trends in counts or numerical values over
time. Because date and time information is continuous categorical data (expressed as a
range of values), points are plotted along the x-axis and connected by a continuous line.

```python
# Time Series Plot
plt.figure(figsize=(8,6))
ts = pd.Series(np.random.randn(100), index = pd.date_range('9/1/2022', periods = 100))
# Return the cumulative sum of the elements.
ts.plot()
ts = ts.cumsum()
ts.plot()
plt.show()
```
