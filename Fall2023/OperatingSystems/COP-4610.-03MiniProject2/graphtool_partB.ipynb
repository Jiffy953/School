```python
import matplotlib.pyplot as plt

# Data from your priority test
pid = [1, 2, 3, 4, 5, 6, 7, 8]
priority = [50, 50, 50, 30, 30, 30, 40, 50]
ticks = [0, 0, 72, 38, 39, 36, 36, 39]

# Plotting the data
width = 0.35  # adjusted the width of the bars for better spacing
ind = range(len(pid))  # the label locations

fig, ax1 = plt.subplots()

# Plotting the 'priority' data
color = 'tab:orange'
ax1.set_xlabel('Process ID')
ax1.set_ylabel('Priority', color=color)
ax1.bar(ind, priority, width, color=color, label='Priority', align='center')
ax1.tick_params(axis='y', labelcolor=color)

# Plotting the 'ticks' data on a secondary y-axis
ax2 = ax1.twinx()
color = 'tab:gray'
ax2.set_ylabel('Ticks', color=color)
ax2.bar([i + width for i in ind], ticks, width, color=color, label='Ticks', align='center')
ax2.tick_params(axis='y', labelcolor=color)

# Correcting the x-axis labels to match the 'pid' values
ax1.set_xticks([i + width/2 for i in ind])
ax1.set_xticklabels(pid)

plt.title('Priority Levels and Ticks by Process ID')
fig.tight_layout()

# Adding a legend to identify the bars
handles, labels = [], []
for ax in fig.axes:
    for h, l in zip(*ax.get_legend_handles_labels()):
        handles.append(h)
        labels.append(l)
plt.legend(handles, labels, loc='upper left')

plt.show()
```
