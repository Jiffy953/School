```python
import matplotlib.pyplot as plt

# Data
pid = [1, 2, 3, 4, 5, 6, 7, 8]
tickets = [1, 1, 1, 10, 20, 30, 40, 50]
ticks = [35, 29, 28, 64, 95, 103, 110, 123]

# Plotting the data
width = 0.35  # the width of the bars
ind = range(len(pid))  # the label locations

fig, ax1 = plt.subplots()

# Plotting the 'tickets' data
color = 'tab:orange'
ax1.set_xlabel('Process ID')
ax1.set_ylabel('Tickets', color=color)
ax1.bar(ind, tickets, width, color=color, label='Tickets')
ax1.tick_params(axis='y', labelcolor=color)

# Plotting the 'ticks' data on a secondary y-axis
ax2 = ax1.twinx()
color = 'tab:gray'
ax2.set_ylabel('Ticks', color=color)
ax2.bar([i + width for i in ind], ticks, width, color=color, label='Ticks')
ax2.tick_params(axis='y', labelcolor=color)

# Correcting the x-axis labels to match the 'pid' values
ax1.set_xticks([i + width/2 for i in ind])
ax1.set_xticklabels(pid)  # Set the x-ticks to the 'pid' list

plt.title('Tickets and Ticks by Process ID')
fig.tight_layout()

# Optionally add a legend if you want to identify the bars
handles, labels = [], []
for ax in fig.axes:
    for h, l in zip(*ax.get_legend_handles_labels()):
        handles.append(h)
        labels.append(l)
plt.legend(handles, labels, loc='upper left')

plt.show()
```
